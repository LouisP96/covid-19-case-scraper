```python
import requests
import re
from bs4 import BeautifulSoup

# Source website
URL = "https://www.worldometers.info/coronavirus/"

page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

results = soup.find(id="main_table_countries_today")

# print(results.prettify())
```

```python
# List of available countries
print("List of available countries:")
for i in results.find_all("a", class_="mt_a"):
    print(i.contents[0])
```

```python
# Enter country below
country = 'UK'

try:
    country_results = results.find("a", string=country).parent.parent
    try:
        country_C19_cases = country_results.find_all("td", style="font-weight: bold; text-align:right;background-color:#FFEEAA;")[0]
    except:
        country_C19_cases = 0
    try:
        country_C19_deaths = country_results.find_all("td", style = re.compile("red"))[0]
    except:
        country_C19_deaths = 0 
        
    country_C19_total_cases = country_results.find_all("td", style = re.compile("right"))[0]
    country_C19_total_deaths = country_results.find_all("td", style = re.compile("right"))[2]

    try:
        print(f"Daily {country} COVID-19 cases:", country_C19_cases.contents[0])
        print(f"Daily {country} COVID-19 deaths:", country_C19_deaths.contents[0], "\n")
    except:
        print(f"Daily {country} COVID-19 cases: 0 (or not yet updated today)")
        print(f"Daily {country} COVID-19 deaths: 0 (or not yet updated today)")

    print(f"Total {country} COVID-19 cases:", country_C19_total_cases.contents[0])
    print(f"Total {country} COVID-19 deaths:", country_C19_total_deaths.contents[0])
except:
    print("Incorrect country name")
    
# To do:
# World and continent values using below cells
```
